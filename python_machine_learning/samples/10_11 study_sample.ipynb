```python
from sklearn import datasets
import numpy as np

# Irisデータセットをロード
iris = datasets.load_iris()

# 3, 4列目の特徴量を抽出('petal length (cm)', 'petal width (cm)')
x = iris.data[:, [2, 3]]

# クラスラベルを取得
y = iris.target
```

```python
#print(iris)
for x_item, y_item in zip(x, y):
    print(y_item, x_item)
```
